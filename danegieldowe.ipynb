```python
df_uber["average"]=(df_uber.Open +df_uber.Close)/2.0
df_uber.head()
```

```python
df_uber=df.sort_index()
df_uber
```

```python
df_uber["Close_shift"]= df_uber.close.shift(1)
df_uber.head()
```

```python
df_uber["daily_change"]= df_uber.close/df_uber.Close_shift - 1
df_uber.head()
```
