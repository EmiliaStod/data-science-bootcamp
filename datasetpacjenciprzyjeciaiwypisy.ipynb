```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generate random IDs
ids = np.arange(1, 101)

# Generate random creation dates within a range
start_date = np.datetime64('2020-01-01')
end_date = np.datetime64('2023-12-31')
creation_dates = [np.random.choice(pd.date_range(start_date, end_date)) for _ in range(100)]

# Generate random last update dates after creation dates
last_update_dates = [date + np.random.randint(1, 365) * np.timedelta64(1, 'D') for date in creation_dates]

# Generate random accepted dates after creation dates
accepted_dates = [date + np.random.randint(1, 365) * np.timedelta64(1, 'D') for date in creation_dates]

# Create DataFrame
data = {
    "id": ids,
    "creation date": creation_dates,
    "last update date": last_update_dates,
    "accepted date": accepted_dates
}

df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())

# Save DataFrame to a CSV file
df.to_csv("inout_dataset.csv", index=False)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Wygenerowanie DataFrame (używając kodu, który podałeś)

# Zliczenie liczby osób przyjętych w każdym roku
creation_dates_count = df['creation date'].dt.year.value_counts().sort_index()

# Zliczenie liczby osób, których ostatnia aktualizacja miała miejsce w każdym roku
last_update_dates_count = df['last update date'].dt.year.value_counts().sort_index()

# Utworzenie DataFrame z danymi do wykresu
data_plot = pd.DataFrame({
    'Przyjęcia': creation_dates_count,
    'Wypisy': last_update_dates_count
})

# Narysowanie wykresu słupkowego
data_plot.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Rok')
plt.ylabel('Liczba pacjentów')
plt.title('Liczba pacjentów w każdym roku przyjętych i wypisanych')
plt.show()


```
