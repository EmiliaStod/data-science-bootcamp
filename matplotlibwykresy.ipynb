```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fig=plt.figure()
fig.show()
```

wykres liniowy

```python
sns.set()
_=plt.plot([0,1],[0,3])
```

```python
_=plt.plot([0,1],[0,3],label='line',color='green')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.title("wykres")
_=plt.legend()
```

```python
x=np.arange(-3,3,0.1)
x_sin=np.sin(x)
x_cos=np.cos(x)
plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

tworzenie wielu wykresow

```python
plt.subplot(121)#tworzy okno z jednym wierszem dwoma kolumnami i pierwszy wykres
plt.plot([0,1,2],[2,3,4])
plt.subplot(122)
plt.plot([0,1,2],[2,3,4])
```

```python
fig=plt.figure(figsize=(12,4))#tutaj regulujemy wielkosc okna wykresow

plt.subplot(131)
plt.plot([0,1,2],[2,3,4])
plt.subplot(132)
plt.plot([0,1,2],[2,3,4],'r--')
plt.subplot(133)
plt.plot([0,1,2],[2,3,4],'ro')
```
